```python
import pandas as pd
from datetime import datetime
import os

# Definir las rutas de los archivos
ruta_clientes = './datos/clientes.csv'
ruta_productos = './datos/productos.csv'
ruta_facturas = './datos/facturas.csv'

# Cargar los datos en DataFrames
df_clientes = pd.read_csv(ruta_clientes)

# Verificar si el archivo productos existe
if os.path.exists(ruta_productos):
      df_productos = pd.read_csv(ruta_productos)
else:
      # Crear un DataFrame vacío con las columnas 'nombre' y 'precio'
      df_productos = pd.DataFrame(columns=['nombre', 'precio'])

# Verificar el archivo facturas

if os.path.exists(ruta_facturas):
      df_facturas = pd.read_csv(ruta_facturas)
else:
      # Crear un DataFrame vacío con las columnas 'cliente', 'producto', 'precio' y 'fecha'
      df_facturas = pd.DataFrame(columns=['cliente', 'producto', 'precio', 'fecha'])


```

```python
import os

def ingresar_producto():
    ruta_productos = './datos/productos.csv'

    # Verificar si el archivo existe
    if os.path.exists(ruta_productos):
        df_productos = pd.read_csv(ruta_productos)
    else:
        # Crear un DataFrame vacío con las columnas 'nombre' y 'precio'
        df_productos = pd.DataFrame(columns=['nombre', 'precio'])

    while True:
        print("**************Menú de productos**************")
        print("1. Nuevo producto - Permite ingresar un nuevo producto al sistema")
        print("2. Guardar archivo - Guarda los productos ingresados en un archivo")
        print("3. Volver - Vuelve al menú principal")
        opcion = input("Seleccione una opción: ")
        os.system('cls' if os.name == 'nt' else 'clear')  # Limpia la consola
        if opcion == "1":
            nombre = input("Ingrese el nombre del producto: ")
            precio = float(input("Ingrese el precio del producto: "))
            df_productos.loc[len(df_productos)] = [nombre, precio]
        elif opcion == "2":
            df_productos.to_csv(ruta_productos, index=False)
            print("Archivo guardado.")
            print(df_productos)
        elif opcion == "3":
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")
            
```

```python


def obtener_cliente(df_clientes):
    while True:
        for i, cliente in enumerate(df_clientes['nombre'], start=1):
            print(f"{i}. {cliente}")
        indice_cliente = int(input("Seleccione el número del cliente: ")) - 1
        if 0 <= indice_cliente < len(df_clientes):
            return df_clientes.iloc[indice_cliente]['nombre']
        else:
            print("Número de cliente no válido. Por favor, intente de nuevo.")

def obtener_producto(df_productos):
    print("**********Productos disponibles:***************")

    # Verificar si el archivo existe
    if os.path.exists(ruta_productos):
        df_productos = pd.read_csv(ruta_productos)
    else:
        # Crear un DataFrame vacío con las columnas 'nombre' y 'precio'
        df_productos = pd.DataFrame(columns=['nombre', 'precio'])
    while True:
        for i, producto in enumerate(df_productos['nombre'], start=1):
            print(f"{i}. {producto}")
        indice_producto = int(input("Seleccione el número del producto: ")) - 1
        if 0 <= indice_producto < len(df_productos):
            return df_productos.iloc[indice_producto]['nombre']
        else:
            print("Número de producto no válido. Por favor, intente de nuevo.")

def ingresar_factura():
    if os.path.exists(ruta_facturas):
        df_facturas = pd.read_csv(ruta_facturas)
    else:
        df_facturas = pd.DataFrame(columns=['cliente', 'producto', 'precio', 'fecha'])

    while True:
        print('******* MENU FACTURACIÓN *******')
        print("1. Ingresar producto")
        print("2. Guardar y salir")
        opcion = input("Seleccione una opción: ")
        os.system('cls' if os.name == 'nt' else 'clear')
        if opcion == "1":
            print("**********Clientes disponibles:***************")
            nombre_cliente = obtener_cliente(df_clientes)
            producto = obtener_producto(df_productos)
            fecha = datetime.now()
            df_facturas.loc[len(df_facturas)] = [nombre_cliente, producto, fecha]
        elif opcion == "2":
            df_facturas.to_csv(ruta_facturas, index=False)
            print("Factura guardada.")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")
```

```python

def ver_resumen():
    while True:
        print('******* MENU RESUMEN *******')
        print("1. Ver resumen")
        print("2. Guardar resumen")
        print("3. Volver")
        opcion = input("Seleccione una opción: ")
        os.system('cls' if os.name == 'nt' else 'clear')
        if opcion == "1":
            cantidad_ventas = len(df_facturas)
            monto_total = df_facturas.merge(df_productos, left_on='producto', right_on='nombre')['precio'].sum()
            print('******* RESUMEN VENTAS *******')
            print(f"Cantidad de ventas: {cantidad_ventas}")
            print(f"Monto total vendido: {monto_total}")
        elif opcion == "2":
            nombre_archivo = input("Ingrese el nombre del archivo: ")
            with open(nombre_archivo, 'w') as f:
                f.write(f"Cantidad de ventas: {cantidad_ventas}\n")
                f.write(f"Monto total vendido: {monto_total}\n")
            print("Resumen guardado.")
        elif opcion == "3":
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

```

```python
def ver_detalle():
    while True:
        print('******* MENU DETALLE *******')
        print("1. Ventas por usuario")
        print("2. Ventas por producto")
        print("3. Ventas diarias")
        print("4. Guardar vista")
        print("5. Volver")
        opcion = input("Seleccione una opción: ")
        if opcion == "1":
            # contar y mostrar las ventas por usuario
            print('-----------------DETALLE VENTAS POR USUARIO-----------------')
            ventas_por_usuario = df_facturas['nombre_cliente'].value_counts()
            ventas_por_usuario_df = ventas_por_usuario.reset_index()
            ventas_por_usuario_df.columns = ['nombre_cliente', 'total de ventas']
            print(ventas_por_usuario_df)
            print('---------------------------------------------------------------')
        elif opcion == "2":
            # contar y mostrar las ventas por producto
            print('-----------------DETALLE VENTAS POR PRODUCTO-----------------')
            ventas_por_producto = df_facturas['producto'].value_counts()
            ventas_por_producto_df = ventas_por_producto.reset_index()
            ventas_por_producto_df.columns = ['producto', 'total de ventas']
            print(ventas_por_producto_df)
            print('---------------------------------------------------------------')
        elif opcion == "3":
            # contar y mostrar las ventas diarias
            print('-----------------DETALLE VENTAS DIARIAS-----------------')
            df_facturas['fecha'] = pd.to_datetime(df_facturas['fecha'])
            ventas_diarias = df_facturas['fecha'].dt.date.value_counts()
            ventas_diarias_df = ventas_diarias.reset_index()
            ventas_diarias_df.columns = ['fecha', 'total de ventas']
            print(ventas_diarias_df)
            print('---------------------------------------------------------------')
        elif opcion == "4":
            nombre_archivo = input("Ingrese el nombre del archivo: ")
            with open(nombre_archivo, 'w') as f:
                if 'ventas_por_usuario' in locals():
                    f.write("Ventas por usuario:\n")
                    f.write(str(ventas_por_usuario))
                    f.write("\n")
                if 'ventas_por_producto' in locals():
                    f.write("Ventas por producto:\n")
                    f.write(str(ventas_por_producto))
                    f.write("\n")
                if 'ventas_diarias' in locals():
                    f.write("Ventas diarias:\n")
                    f.write(str(ventas_diarias))
                    f.write("\n")
            print("Vista guardada.")
        elif opcion == "5":
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")
            
```

```python

import pandas as pd
import os

def borrar_datos():
    while True:
        print('******* MENU BORRAR DATOS *******')
        print("1. Borrar productos")
        print("2. Borrar facturas")
        print("3. Borrar todos")
        print("4. Volver")
        opcion = input("Seleccione una opción: ")
        os.system('cls' if os.name == 'nt' else 'clear')
        if opcion == "1":
            if os.path.exists(ruta_productos):
                os.remove(ruta_productos)
                print("Archivo de productos borrado.")
                # Restablecer el DataFrame de productos
                df_productos = pd.DataFrame(columns=['nombre', 'precio'])
                print(df_productos)
            else:
                print("El archivo de productos no existe.")
        elif opcion == "2":
            if os.path.exists(ruta_facturas):
                os.remove(ruta_facturas)
                print("Archivo de facturas borrado.")
                # Restablecer el DataFrame de facturas
                df_facturas = pd.DataFrame(columns=['factura', 'producto', 'cantidad'])
            else:
                print("El archivo de facturas no existe.")
        elif opcion == "3":
            if os.path.exists(ruta_productos):
                os.remove(ruta_productos)
                print("Archivo de productos borrado.")
                df_productos = pd.DataFrame(columns=['nombre', 'precio'])
            if os.path.exists(ruta_facturas):
                os.remove(ruta_facturas)
                print("Archivo de facturas borrado.")
                df_facturas = pd.DataFrame(columns=['factura', 'producto', 'cantidad'])
        elif opcion == "4":
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

```

```python
import os

def main():
    while True:
        print("**************MENU******************")
        print("1. Ingresar producto - Permite ingresar un nuevo producto al sistema")
        print("2. Ingresar factura - Permite ingresar una nueva factura al sistema")
        print("3. Ver resumen - Muestra un resumen de las ventas")
        print("4. Ver detalle - Muestra un detalle de las ventas")
        print("5. Borrar datos - Borra todos los datos del sistema")
        print("6. Salir - Sale del programa")
        opcion = input("Seleccione una opción: ")
        os.system('cls' if os.name == 'nt' else 'clear')  # Limpia la consola
        if opcion == "1":
            ingresar_producto()
        elif opcion == "2":
            ingresar_factura()
        elif opcion == "3":
            ver_resumen()
        elif opcion == "4":
            ver_detalle()
        elif opcion == "5":
            borrar_datos()
        elif opcion == "6":
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

if __name__ == "__main__":
    main()
```
