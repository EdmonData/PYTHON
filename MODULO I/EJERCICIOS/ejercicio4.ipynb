1. Escriba un programa en Python para crear una clase que represente
un círculo. Incluya métodos para calcular su área y perímetro.

```python
#importamos la librería math para poder usar la constante pi
import math

# Definir la clase Circulo
class Circulo:
    def __init__(self, radio):
        self.radio = radio

    def calcular_area(self):
        return math.pi * self.radio ** 2

    def calcular_perimetro(self):
        return 2 * math.pi * self.radio

# Instancia la clase Circulo con un radio de 5
circulo = Circulo(5)

# Calcular e imprimir el área y el perímetro del círculo
print("El área del círculo es: ", circulo.calcular_area())
print("El perímetro del círculo es: ", circulo.calcular_perimetro())
```

2. Escriba un programa en Python para crear una clase ‘persona’. Incluya
atributos tales como nombre, país y fecha de nacimiento. Implemente
un método para determinar la edad de la persona.

```python

from datetime import datetime

# Definir la clase Persona
class Persona:
    def __init__(self, nombre, pais, fecha_nacimiento):
        self.nombre = nombre
        self.pais = pais
        self.fecha_nacimiento = datetime.strptime(fecha_nacimiento, "%d/%m/%Y")

    def calcular_edad(self):
        hoy = datetime.today()
        return hoy.year - self.fecha_nacimiento.year - ((hoy.month, hoy.day) < (self.fecha_nacimiento.month, self.fecha_nacimiento.day))

persona = Persona("Juan", "Colombia", "01/01/1990")

# Imprimir el nombre y el país de la persona
print("Nombre: ", persona.nombre)
print("País: ", persona.pais)

# Calcular e imprimir la edad de la persona
print("La edad de la persona es: ", persona.calcular_edad())
```

3. Escriba un programa en Python para crear una clase ‘calculadora’.
Incluya métodos de operaciones aritméticas básicas: suma, resta,
multiplicación y división.

```python
# Definir la clase Calculadora
class Calculadora:
    def sumar(self, a, b):
        return a + b

    def restar(self, a, b):
        return a - b

    def multiplicar(self, a, b):
        return a * b

    def dividir(self, a, b):
        if b != 0:
            return a / b
        else:
            return "Error: División por cero"

# Instancia la clase Calculadora
calculadora = Calculadora()

# Realizar operaciones matemáticas con la calculadora
print("Suma: ", calculadora.sumar(5, 3))
print("Resta: ", calculadora.restar(5, 3))
print("Multiplicación: ", calculadora.multiplicar(5, 3))
print("División: ", calculadora.dividir(5, 3))
```

4. Escriba un programa en Python para crear una clase que represente
una forma. Incluya métodos para calcular su área y perímetro.
Implemente subclases para las formas círculo, triángulo y cuadrado.

```python
import math

# Definir la clase Forma
class Forma:
    def calcular_area(self):
        pass

    def calcular_perimetro(self):
        pass

# Definir la clase Circulo
class Circulo(Forma):
    def __init__(self, radio):
        self.radio = radio

    def calcular_area(self):
        return math.pi * self.radio ** 2

    def calcular_perimetro(self):
        return 2 * math.pi * self.radio

# definir la clase Triangulo
class Triangulo(Forma):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        return 0.5 * self.base * self.altura

    def calcular_perimetro(self):
        return self.base + 2 * math.sqrt((self.base / 2) ** 2 + self.altura ** 2)

# definir la clase Cuadrado
class Cuadrado(Forma):
    def __init__(self, lado):
        self.lado = lado

    def calcular_area(self):
        return self.lado ** 2

    def calcular_perimetro(self):
        return 4 * self.lado

# Instancia la clase Circulo con un radio de 5
circulo = Circulo(5)

# Calcular e imprimir el área y el perímetro del círculo
print("El área del círculo es: ", circulo.calcular_area())
print("El perímetro del círculo es: ", circulo.calcular_perimetro())

# Instancia la clase Triangulo con una base de 5 y una altura de 3
triangulo = Triangulo(5, 3)

# Calcular e imprimir el área y el perímetro del triángulo
print("El área del triángulo es: ", triangulo.calcular_area())
print("El perímetro del triángulo es: ", triangulo.calcular_perimetro())

# Instancia la clase Cuadrado con un lado de 5
cuadrado = Cuadrado(5)

# Calcular e imprimir el área y el perímetro del cuadrado
print("El área del cuadrado es: ", cuadrado.calcular_area())
print("El perímetro del cuadrado es: ", cuadrado.calcular_perimetro())






```

5. Escriba un programa en Python para crear una clase que represente
un carro de compras. Incluya métodos para agregar y quitar ítems, y
para calcular el precio total.

```python
# Definir la clase CarroDeCompras
class CarroDeCompras:
    def __init__(self):
        self.items = {}

    def agregar_item(self, nombre, precio):
        self.items[nombre] = precio

    def quitar_item(self, nombre):
        if nombre in self.items:
            del self.items[nombre]
        else:
            print("El ítem no está en el carro de compras")

    def calcular_precio_total(self):
        return sum(self.items.values())

# Instancia la clase CarroDeCompras
carro_de_compras = CarroDeCompras()

# Agregar ítems al carro de compras
carro_de_compras.agregar_item("Camisa", 20)
carro_de_compras.agregar_item("Pantalón", 40)
carro_de_compras.agregar_item("Zapatos", 50)

# Imprimir los ítems del carro de compras
print("Ítems del carro de compras: ", carro_de_compras.items)

# Calcular e imprimir el precio total del carro de compras
print("Precio total del carro de compras: ", carro_de_compras.calcular_precio_total())

# Quitar un ítem del carro de compras
carro_de_compras.quitar_item("Pantalón")

# Imprimir los ítems del carro de compras
print("Ítems del carro de compras: ", carro_de_compras.items)

# Calcular e imprimir el precio total del carro de compras
print("Precio total del carro de compras: ", carro_de_compras.calcular_precio_total())



```
