A    Aprendizaje supervisado

RESP 1. A Aprendizaje supervisado: es un enfoque de machine learning que utiliza conjuntos de datos etiquetados y resultados correctos para entrenar algoritmos.
Estos algoritmos clasifican datos o predicen resultados con precisión.

Por ejemplo, se usa para clasificar spam o predecir precios de viviendas.

B   Aprendizaje sin supervisión

RESP 1. B Aprendizaje sin supervisión es un enfoque de machine learning que analiza y agrupa datos sin etiquetas.
Descubre patrones ocultos sin intervención humana. Útil para exploración de datos y segmentación de clientes.
Por ejemplo, se usa para encontrar grupos similares en datos no clasificados.

C   Aprendizaje por refuerzo

RESP 1. C El aprendizaje por refuerzo es una técnica de machine learning que entrena al software para tomar decisiones y lograr los mejores resultados.
Imita el proceso de aprendizaje por ensayo y error que los humanos utilizan para alcanzar objetivos.
Las acciones que se acercan al objetivo se refuerzan, mientras que las que se apartan se ignoran.
Los algoritmos de RL aprenden de la retroalimentación y descubren las mejores rutas para lograr resultados finales

D   Algoritmos de regresión

RESP 1. D Los algoritmos de regresión son un conjunto de técnicas utilizadas en aprendizaje automático que se enfocan en predecir o estimar valores continuos.
Estos algoritmos se basan en la relación entre una variable dependiente y una o más variables independientes.
Por ejemplo, se utilizan para predecir precios de viviendas o analizar tendencias financieras.

E   Algoritmos de agrupación

RESP 1. E Los algoritmos de agrupación son un conjunto de técnicas utilizadas en aprendizaje automático que se enfocan en predecir o estimar valores continuos.
Estos algoritmos se basan en la relación entre una variable dependiente y una o más variables independientes.
Por ejemplo, se utilizan para predecir precios de viviendas o analizar tendencias financieras.

F   Algoritmos de Aprendizaje Profundo

RESP1. F Los algoritmos de Aprendizaje Profundo son un conjunto de técnicas en Inteligencia Artificial que se inspiran en el cerebro humano.
Estos modelos pueden reconocer patrones complejos en imágenes, texto y sonidos para generar predicciones precisas.
Se utilizan en asistentes digitales, detección de fraudes, reconocimiento facial y más
