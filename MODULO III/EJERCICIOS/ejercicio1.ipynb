Caso:
En una clínica infantil se ha ido anotando, durante un mes, el número de metros que cada niño anda, seguido y sin caerse, el primer día que comienza
a caminar, obteniéndose la tabla de información adjunta:
Metros 1 2 3 4 5 6 7 8
Niños 2 6 10 5 10 3 2 2

Se pide:
a) Tabla  de  Frecuencia.  Diagrama  de  Barras  para  las  frecuencias absolutas, relativas y acumuladas.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear DataFrame
data = {'Metros': [1, 2, 3, 4, 5, 6, 7, 8],
        'Niños': [2, 6, 10, 5, 10, 3, 2, 2]}
df = pd.DataFrame(data)

# Crear los datos a partir de la frecuencia
metros = np.repeat(df['Metros'], df['Niños'])
# Crear la matriz
matriz = np.array(metros)

print(matriz)

```

```python
# Calcular las frecuencias absolutas
df['Frecuencia_Absoluta'] = df['Niños']

# Calcular las frecuencias relativas
df['Frecuencia_Relativa'] = df['Niños'] / df['Niños'].sum()

# Calcular las frecuencias acumuladas
df['Frecuencia_Acumulada'] = df['Frecuencia_Absoluta'].cumsum()

# Mostrar la tabla de frecuencia
print(df.head(10))

# Crear el diagrama de barras para las frecuencias absolutas
plt.bar(df['Metros'], df['Frecuencia_Absoluta'])
plt.title('Frecuencias Absolutas')
plt.xlabel('Metros')
plt.ylabel('Frecuencia')
plt.show()

# Crear el diagrama de barras para las frecuencias relativas
plt.bar(df['Metros'], df['Frecuencia_Relativa'])
plt.title('Frecuencias Relativas')
plt.xlabel('Metros')
plt.ylabel('Frecuencia')
plt.show()

# Crear el diagrama de barras para las frecuencias acumuladas
plt.bar(df['Metros'], df['Frecuencia_Acumulada'])
plt.title('Frecuencias Acumuladas')
plt.xlabel('Metros')
plt.ylabel('Frecuencia')
plt.show()

# Crear diagrama de barras con la frecuencia relativa en un segundo eje y ademas que muestre un titulo y todas las leyendas
ax = df[['Frecuencia_Absoluta', 'Frecuencia_Acumulada']].plot(kind='bar', figsize=(12, 6))
df['Frecuencia_Relativa'].plot(kind='bar', secondary_y=True, ax=ax, alpha=0.5, color='g', label='Frecuencia Relativa')
ax.set_title('Tu título aquí')
ax.set_xlabel('Metros')
ax.set_ylabel('Frecuencia Absoluta y Acumulada')
ax.legend(loc='best')
```

b) Mediana, media aritmética, moda y cuartiles de los metros


```python


import scipy.stats as stats

# Calcular la mediana
mediana = np.median(matriz)

# Calcular la media aritmética
media = np.mean(matriz)

# Calcular la moda
moda = stats.mode(matriz)

# Calcular los cuartiles
cuartil1 = np.percentile(matriz, 25)
cuartil2 = np.percentile(matriz, 50)
cuartil3 = np.percentile(matriz, 75)

print(f"La mediana es: {mediana} metros")
print(f"La media aritmética es: {media} metros")
#print(f"La moda es: {moda.mode[0]} metros")
print(f"El primer cuartil es: {cuartil1} metros")
print(f"El segundo cuartil es: {cuartil2} metros")
print(f"El tercer cuartil es: {cuartil3} metros")
```

c)  Varianza y desviación Estándar.


```python

# Calcular la varianza
varianza = np.var(matriz)
print(f"Varianza: {varianza}")

# Calcular la desviación estándar
desviacion_estandar = np.std(matriz)
print(f"Desviación estándar: {desviacion_estandar}")
```

```python
# Calcular los percentiles 25 y 75
percentil_25 = np.percentile(matriz, 25)
percentil_75 = np.percentile(matriz, 75)

print(f"El 75% de las observaciones se encuentra entre {percentil_25} y {percentil_75} metros")
```
