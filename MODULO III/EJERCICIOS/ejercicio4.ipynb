b) Cargar el dataset de titanic, se utilizará en los siguientes ejercicios.


```python
# Importar seaborn
import seaborn as sns

# Cargar el dataset de Titanic
df = sns.load_dataset('titanic')

# Mostrar las primeras 5 filas del dataframe
print(df.head())
```

c)  Confeccione un gráfico de barras para visualizar hombres y mujeres que han sobrevivido

```python
# Importar seaborn
import seaborn as sns

# Importar matplotlib
import matplotlib.pyplot as plt

# Reemplazar 'female' y 'male' por 'mujer' y 'hombre'
df['sex'] = df['sex'].replace({'female': 'mujer', 'male': 'hombre'})

# Crear un gráfico de barras de 'sex' y 'survived'
grafico = sns.countplot(x='sex', hue='survived', data=df)

# Agregar títulos y etiquetas
plt.title('Sobrevivientes por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de sobrevivientes')

# Cambiar la leyenda
grafico.legend(['Muertos', 'Sobrevivientes'])

# Mostrar el gráfico
plt.show()
```

d) Crear un gráfico de distribución de edades por clase de pasajero .

```python
# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un histograma de 'age' para cada 'class'
for clase in df['class'].unique():
    plt.hist(df[df['class'] == clase]['age'].dropna(), bins=30, alpha=0.5, label=str(clase))

# Agregar títulos y etiquetas
plt.title('Distribución de edades por clase de pasajero')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend(title='Clase')

# Mostrar el gráfico
plt.show()
```

e) Diagrama de Caja (Boxplot) de Tarifas por Puerto de Embarque.

```python
# Importar seaborn
import seaborn as sns

# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un boxplot de 'fare' para cada 'embark_town'
sns.boxplot(x='embark_town', y='fare', data=df)

# Agregar títulos y etiquetas
plt.title('Tarifas por puerto de embarque')
plt.xlabel('Puerto de embarque')
plt.ylabel('Tarifa')

# Mostrar el gráfico
plt.show()
```

f)  Gráfico de Violín de Edades por Supervivencia.

```python
# Importar seaborn
import seaborn as sns

# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un gráfico de violín de 'age' para cada 'survived'
sns.violinplot(x='survived', y='age', data=df)

# Agregar títulos y etiquetas
plt.title('Edades por supervivencia')
plt.xlabel('Supervivencia')
plt.ylabel('Edad')

# Mostrar el gráfico
plt.show()
```

g) Diagrama de dispersión de Tarifas vs. Edades por Supervivencia.

```python
# Importar seaborn
import seaborn as sns

# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un diagrama de dispersión de 'fare' vs 'age' coloreado por 'survived'
sns.scatterplot(x='age', y='fare', hue='survived', data=df)

# Agregar títulos y etiquetas
plt.title('Tarifas vs. Edades por Supervivencia')
plt.xlabel('Edad')
plt.ylabel('Tarifa')

# Mostrar el gráfico
plt.show()
```

h) Crear un gráficos de pilas agrupadas según supervivencia vs clase, agrupadas por puerto de embarque.

```python
# Importar pandas y matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe pivotado para el gráfico
pivot_df = df.pivot_table(values='survived', index=['embark_town', 'class'], aggfunc='sum', observed=False).unstack()

# Crear un gráfico de barras apiladas
pivot_df.plot(kind='bar', stacked=True)

# Agregar títulos y etiquetas
plt.title('Supervivencia vs Clase, agrupadas por Puerto de Embarque')
plt.xlabel('Puerto de Embarque, Clase')
plt.ylabel('Supervivencia')

# Mostrar el gráfico
plt.show()
```

i) pairplot para mostrar gráficos de dispersión y histogramas de las variables 'Age', 'Fare' y 'Pclass' del Titanic, diferenciando entre los pasajeros que sobrevivieron y los que no.

```python
# Importar seaborn
import seaborn as sns

# Crear un pairplot de 'age', 'fare' y 'pclass', coloreado por 'survived'
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']].dropna(), hue='survived')

# Mostrar el gráfico
plt.show()
```
