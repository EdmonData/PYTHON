o Dispersión:  sirve  para  representar  la  relación  entre  dos variables. Para nuestro caso se desea ver la relación entre índice de crimen y el valor mediano de las casas. En otras palabras, ¿el índice de crimen afecta el valor medio de las casas?
o Barras: Un gráfico de barras es útil para comparar una variable entre distintos grupos o categorías. Por ejemplo, si quisiéramos observar el valor medio de cada ciudad. Para nuestro caso se desea agrupar por ciudad y toma la media del valor mediano. Gráfica las 10 primeras ciudades.
o Cajas: Los diagramas de cajas son útiles para representar grupos de datos y compararlos entre ellos. Otro uso importante es que nos permiten identificar de forma sencilla si una variable tiene muchos outliers (valores atípicos) esto es, elementos que se alejan de los valores frecuentes de dicha variable. Para nuestro caso, se desea ver los valores atípicos de índice de crimen en los diferentes cuantiles de valor mediano.

```python
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv('datos/casasboston.csv')

# Renombrar las columnas
df.columns = ['CIUDAD', 'LONGITUD', 'LATITUD', 'VALOR_MEDIO_VIVIENDAS_OCUPADAS', 'TAZA_CRIMENES', 'PORCION_DE_TERRENO', 'INDUSTRALIZACION', 'VARIABLE_CHARLES_R', 'CONCENT_DE_OXIDOS', 'MEDIA_HABITACIONES_X_VIVIENDA', 'ANTIGUEDAD', 'DISTANCIA_CENTROS_EMPLEOS', 'INDICE_DISTANCIA_CARRETERAS', 'IMPUESTO', 'PROPORCION_ALUMNO_MAESTRO', 'PROPORCION_AFROAMERICANOS', 'PORCENTAJE_POB_ESTATUS_BAJO']

# Mostrar las primeras filas del DataFrame
print(df.head(2000))
print(df.columns)

#renombrar las columna en nombres en español

```

1 Crear 2 gráficos de Histograma

```python
# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un histograma de 'VALOR_MEDIO_VIVIENDAS_OCUPADAS'
plt.hist(df['ANTIGUEDAD'], bins=30, color='blue', edgecolor='black')

# Agregar títulos y etiquetas
plt.title('Histograma de VALOR MEDIO VIVIENDASOCUPADAS')
plt.xlabel('VALOR_MEDIO_VIVIENDAS_OCUPADAS')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```

```python
# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un histograma de 'CONCENT_DE_OXIDOS'
plt.hist(df['CONCENT_DE_OXIDOS'], bins=30, color='green', edgecolor='black')

# Agregar títulos y etiquetas
plt.title('Histograma de CONCENT_DE_OXIDOS')
plt.xlabel('CONCENT_DE_OXIDOS')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```

2) crear 2 gráficos de dispersion

```python
# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un gráfico de dispersión entre 'PROPORCION_AFROAMERICANOS' y 'TAZA_CRIMENES'
plt.scatter(df['PROPORCION_AFROAMERICANOS'], df['TAZA_CRIMENES'], color='orange')

# Agregar títulos y etiquetas
plt.title('Gráfico de dispersión entre PROPORCION_AFROAMERICANOS y TAZA_CRIMENES')
plt.xlabel('PROPORCION_AFROAMERICANOS')
plt.ylabel('TAZA_CRIMENES')

# Mostrar el gráfico
plt.show()
```

```python
# Importar matplotlib
import matplotlib.pyplot as plt

# Crear un gráfico de dispersión entre 'VALOR_MEDIO_VIVIENDAS_OCUPADAS' y 'TAZA_CRIMENES'
plt.scatter(df['VALOR_MEDIO_VIVIENDAS_OCUPADAS'], df['TAZA_CRIMENES'], color='blue')

# Agregar títulos y etiquetas
plt.title('Gráfico de dispersión entre VALOR_MEDIO_VIVIENDAS_OCUPADAS y TAZA_CRIMENES')
plt.xlabel('VALOR_MEDIO_VIVIENDAS_OCUPADAS')
plt.ylabel('TAZA_CRIMENES')

# Mostrar el gráfico
plt.show()
```
